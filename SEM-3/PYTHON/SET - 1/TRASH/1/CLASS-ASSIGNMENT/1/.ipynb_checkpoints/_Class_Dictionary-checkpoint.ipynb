##### 1. Sort Dictionary

```python
# 1. Sort Dictionary
d = dict()
for x in sorted(_):
    d[x] = _[x]
print(d)
d = dict()
for x in sorted(_,reverse=True):
    d[x] = _[x]
print(d)

####################################
a = {1:'x',
    2:'b',
    3:'a',
    4:'w'}

item = list(a.values())
item.sort()
i=0
for x in a.keys():
    a[x] = item[i]
    i+=1

print(a)

####################################
a = {1:'x',
    2:'b',
    3:'a',
    4:'w'}
key = list(a.keys())
key.sort()
b = dict()
i=0
for x,y in a.items():
    b[x] = key[i]
    i+=1
print(b)

####################################
# Sort by Keys
a = {3:'x',
    1:'b',
    4:'a',
    2:'w'}
b = dict()
for x in sorted(a):
    b[x] = a[x]
print(b)

# Sort by Values
a = {1:'x',
    2:'b',
    3:'a',
    4:'w'}
print("Original :",a)
key = list(a.keys())
print("keys :",key)
value = list(a.values())
print("ori_value :",value)
ori_value = value.copy()
value.sort()
print("modified :",value)

b=dict()

for x in value:
    b[key[ori_value.index(x)]] = x
print("Sorted :",b)
```

##### 2. Sum of items in the dictionary

```python
# 2. Sum of items in the dictionary :
a = {1:'x',
    2:'b',
    3:'a',
    4:'w'}
str_val = ""
int_key = 0
for x,y in a.items():
    str_val+=y
    int_key+=x
print("Keys: ",str_val,"..... Value: ",int_key)
```

##### 3. Extract Key's Value, if Key Present in List and Dictionary

```python
# 3. Extract Key's Value, if Key Present in List and Dictionary
def extract(ls,dt):
    ret = []
    for x in ls:
        if dt.get(x) is not None:
            ret.append(dt.get(x))
    return ret

ls =[1,2,3,4,5]
dt = {1:'x',
    2:'b',
    3:'a',
    4:'w'}

print(extract(ls,dt))
```

##### 4. Count the Number of Keys, Value, and Key-Value Pair

```python

def count(dt):
    key_count = len(dt.keys())
    value_count = len(dt.values())
    key_value_count = len(dt)
    return key_count, value_count, key_value_count

dt = {1:'x',
    2:'b',
    3:'a',
    4:'w'}

print(count(dt))
```
