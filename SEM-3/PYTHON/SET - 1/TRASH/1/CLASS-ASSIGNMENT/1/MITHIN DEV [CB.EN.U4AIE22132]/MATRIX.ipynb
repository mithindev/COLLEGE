### Initializing a Matrix

```python
class Matrix:
    def __init__(self):
        self.matrix = []
        self.row = int(input("Enter the Number of Rows : "))
        self.col = int(input("Enter the Number of Columns : "))
        for row in range(self.row):
            emp = []
            for col in range(self.col):
                emp.append(int(input(f"Enter the value for Element {row+1}{col+1} : ")))
            self.matrix.append(emp)
        self.display()
    
    def display(self, mat=None):
        if mat is None:
            print(self.__str__())
        else:
            for row in range(len(mat)):
                for col in range(len(mat[row])):
                    print(mat[row][col], end=" ")
                print("")
    
    def __str__(self):
        string = ""
        for row in range(self.row):
            for col in range(self.col):
                string+=str(self.matrix[row][col])+" "
            string.strip()
            string+='\n'
        return string

    
    def add(self, mat2):
        mat = []
        for row in range(self.row):
            a = []
            for col in range(self.col):
                a.append(self.matrix[row][col] + mat2[row][col])
            mat.append(a)
        self.display(mat)
        return mat
    
    def transpose(self):
        mat = []
        for row in range(self.row):
            emp = []
            for col in range(self.col):
                emp.append(self.matrix[col][row])
            mat.append(emp)
        self.matrix = mat
        self.display()
        return mat
    
    def symmetric(self):
        for x in range(self.row):
            for y in range(self.col):
                if self.matrix[x][y] != self.matrix[y][x]:
                    return False
        
        return True

mat1 = Matrix()
mat2 = Matrix()
add_mat = mat1.add(mat2.matrix)
trans = mat1.transpose()
```
