### Count the occurance of 'm' in Malayalam

```python
# Count the occurance of 'm' in Malayalam
def count(word,letter):
    count = 0
    for lt in word:
        if lt==letter:
            count+=1
    return count

def count1(word,letter):
    return word.count(letter)

count("Malayalam","m")
count1("Malayalam","a")
```

### Palindrome

```python
# Palindrome String 
def palindrome(word):
    word = word.lower()
    for x in range(len(word)//2):
        if word[x]!=word[len(word)-1-x]:
            return False
    return True

def palindrome_1(word):
    return [word[x] for x in range(len(word)-1,-1,-1)] == list(word)

def palindrome_2(word):
    return word[::-1]==word

print(palindrome('Malayalam'))
```

### Frequency

```python
# count frequency
def frequency(word):
    word = word.lower()
    x = dict()
    for y in word:
        if y in ['a','e','i','o','u']:
            x[y] = x.get(y,0) + 1
    return x

print(frequency("Mithin dev"))
```

### Matching Character

```python
# Find matching character
def matching(word1,word2):
    count = 0
    for x in range(min(len(word1),len(word2))):
        if word1[x]==word2[x]:
            count+=1
    return count

def matching_1(word1,word2):
    word1, word2 = word1.lower(), word2.lower()
    arr1 = [0 for x in range(26)]
    arr2 = arr1.copy()
    for x in word1:
        arr1[ord(x)-ord('a')] +=1
    for x in word2:
        arr2[ord(x)-ord('a')] +=1
    
    count = 0
    for x in range(26):
        count+= min(arr1[x],arr2[x])
    
    return count
    
```
