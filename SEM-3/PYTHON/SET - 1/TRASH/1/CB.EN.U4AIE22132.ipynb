#### 15. For the given input list, filter all whole elements 12\nthree irrespective of case.
items = ['12\nthree', '12\nThree'l, '12\nthree\n4', '12\nthree’]
Output: ['12\nThree', '12\nthree'l]

```python
import re

list15 = ['12\nthree\n', '12\nThree', '12\nthree\n4', '12\nthree']
listo5 = list(filter(lambda item: re.match(r'^12\nthree$', item, re.IGNORECASE) and item != list15[0], list15))
print(listo5)
```

#### 17. For the given input list, filter all elements starting with h. Additionally, replace e with X for these filtered elements.
