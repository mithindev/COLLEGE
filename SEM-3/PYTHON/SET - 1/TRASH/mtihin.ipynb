```python
import re

pattern = '0xB0'
txt = 'end address: 0xFF,func2 address:0xB0'

ans = re.search(pattern, txt)

print(ans)
```

```python
import re

pattern = '5'
replacement = 'five'
txt = 'they ate 5 apples and 5 oranges'

ans = re.sub(pattern, replacement, txt)

print(ans)
```

```python
import re

pattern = '5'
replacement = 'five'
txt = 'they ate 5 apples and 5 oranges'

ans = re.sub(pattern, replacement, txt, count = 1)

print(ans)
```

```python
import re

text = 'This note should not be NoTeD'

ans = re.sub(r'note', 'X', text, flags = re.IGNORECASE)

print(ans)
```
