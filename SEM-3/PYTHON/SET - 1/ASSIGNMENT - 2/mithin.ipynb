## Q - 7

Write a Python program that matches a string that has an ‘a’ followed by zero or more b's.

```python
import re

pattern = r'ab*'

input_string = input("Enter a string: ")

if re.search(pattern, input_string):
    print("Match found!")
else:
    print("No match found.")

'''
Explaination:

The Python program uses regular expressions with the pattern "ab*" to match a string.
It searches for an 'a' followed by zero or more 'b's.
It prompts the user for input, then checks if the input string contains the specified pattern using re.search().
If a match is found, it prints "Match found!"; otherwise, it prints "No match found."
'''
```

## Q - 8

Write a Python program to find sequences of lowercase letters joined with anunderscore.

```python
import re

input_string = input("Enter a string: ")

pattern = r'[a-z]+_[a-z]+'

matches = re.findall(pattern, input_string)

if matches:
    print("Seq found!")
    for match in matches:
        print(match)
else:
    print("Not found!")

'''
Explainaion:
It prompts the user for input, defines the regular expression pattern '[a-z]+_[a-z]+' to find such sequences, and employs 're.findall()' to locate all matches.
If matches are found, it prints them; otherwise, it indicates that no matching sequences were found.
'''
```

## Q - 9

Write a Python program to match a string that contains only upper and lowercase letters, numbers, and underscores.

```python
import re

input_string = input("Enter a string: ")

pattern = r'^[a-zA-Z0-9_]+$'

if re.match(pattern, input_string):
    print("Match found!")
else:
    print("No match found.")

'''
Explaination:
we use a regular expression pattern '^[a-zA-Z0-9_]+$' to check if the input string contains only upper and lowercase letters, numbers, and underscores. It employs 're.match()' to validate the entire string, printing either a match confirmation or a message indicating the presence of other characters.
'''
```
