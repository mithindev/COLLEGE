#### 1. Find the sum of even numbers upto an positive integer

```python
def sum_evev(n) :
    ans = 0

    for i in range(n + 1):
        if (i % 2 == 0):
            ans += i

    return ans
```

#### 2. Find the factorial of an positive number

```python
def fact(n):
    if (n == 1 or n == 0):
        return 1
    else:
        return n * fact(n - 1)
```

#### 3. Print the multiplication table of a given positive integer

```python
def mult_table (num):
    for i in range(1, 11):
        print(num * i)
```

#### 4. Display the Fibonacci series upto an given number of terms

```python
def fibo(n):
    if n <= 1:
        return n
    else:
        return fibo(n - 1) + fibo(n - 2)
```

#### 5. Check if an given string is an palindrome

```python
def isPalindrome (str):
    s = 0
    e = len(str) - 1

    while (s <= e):
        if (str[s] != str[e]):
            return False
        s += 1
        e -= 1
    return True
```

#### 6. Check whether a given number is prime or not

```python
def isPrime(num):
    i = 2

    while (i*i <= num):
        if (num % i == 0):
            return False
        num += 1

    return True

print(isPrime(3))
```
