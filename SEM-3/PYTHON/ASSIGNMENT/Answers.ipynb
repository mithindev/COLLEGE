```python
def leap(y):
    if (y % 4 == 0 and y % 100 != 0) or (y % 100 == 0 and y % 400 == 0):
        return True
    return False

def dayfinder(y, m):
    i = 1
    days = 0
    while i < y:
        days += 366 if leap(i) else 365
        i += 1
        
    year = {1: 31, 2: 29 if leap(y) else 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
    lst = list(year.values())
    
    for i in range(m-1):
        days += lst[i]
        
    day = (days + 1) % 7
    
    return day

def calPrinter(m, y):
    
    if m > 12 or m < 0:
        print("Please choose a valid month!!")
        return
    
    if y < 0:
        print("Please choose a valid year!!")
        return
    d = dayfinder(y, m)
    
    count = 0
    
    date = 1
    
    year = {1: 31, 2: 29 if leap(y) else 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
    
    print(f"      x|{m}|{y}")
    
    print("Su Mo Tu We Th Fr Sa", end = '\n')
    
    for i in range(d):
        count += 1
        print('  ', end = ' ')
        
    
    for i in range(year.get(m)):
        if count % 7 == 0:
            print('\n', end = '')
            
        print(f"{date:02d}", end = ' ')
        count += 1
        date += 1
        

calPrinter(10, 1971)
```

```python
def permutation (ans, word):
  if len(word) == 0:
    return [ans]
  ch = word[0]
  ans_lst = []
  for i in range(0, len(ans) + 1):
    left = ans[:i]
    right = ans[i:]
    ans_lst += permutation(left + ch + right, word[1:])
  return ans_lst
def generate_permutations(word):
    return permutation("", word)

ans = permutation("", "abc")
print(ans)
```

```python
def lcs(text1: str, text2: str) -> int:
        if len(text1) == 0 or len(text2) == 0:
            return 0
        elif text1[len(text1) - 1] == text2[len(text2) - 1]:
            return 1 + lcs(text1[:len(text1) - 1], text2[:len(text2) - 1])
        else:
            return max(lcs(text1[:len(text1)], text2[:len(text2) - 1]), lcs(text1[:len(text1) - 1], text2[:len(text2)]))

print(lcs('abc', 'abcd'))
```

```python
import random
import string

def password(n):
    letters = string.ascii_letters
    digits = string.digits
    symbols = string.punctuation
    all_characters = letters + digits + symbols
    password = ''.join(random.choice(all_characters) for _ in range(n))#generation of random password
    return password

n = 12 #length of the password
random_password = password(n)#function call
print(random_password)
```

```python
def lastname(a):
    c=a.split()
    d=c[-1]
    return d
def sorter(aa):
    n=len(aa)
    for i in range(len(aa)):
        for j in range(0,n-1-i):
            if lastname(aa[j])>lastname(aa[j+1]):#last name alone is used for comparing
                aa[j],aa[j+1]=aa[j+1],aa[j]#with that relation now we are comparing the originallist 
    return aa

aa= ["John Smith", "Alice Johnson", "Bob Brown", "Eve Adams"]
ccc=sorter(aa) 
print(ccc)
```

```python
def check(left, right):
    while left >= 0 and right < len(a) and a[left] == a[right]:#from center they go left and right and check if the string matches or not in this while loop
        left -= 1
        right += 1
    return a[left + 1:right]#returning from left to right which are explored and found that they are palindrome

def l_pal(a):
    longest = ""
    for i in range(len(a)):
        for j in range(i + 1, len(a)):
            if a[i] == a[j]:#search for center and i found their locations will be sent to call check function for the further process
                palindrome = check(i, j)
                if len(palindrome) > len(longest):
                    longest = palindrome

    return longest
a="abccbassssssss"
l_pal(a)
```

```python
def rom_to_int(s):
    rom={'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
    prev=0
    result=0
    for char in reversed(s):
        current=rom[char]# in each iteration current will have the current iterating letter.
        if(current<prev):
            result=result-current
        else:
            result=result+current
        
        prev=current
    return result

rom_to_int('IV')
```

```python
def diamond(n):
    for i in range(n-1):#upper diamond
        for j in range(n-1-i):
            print(" ",end="")
        for k in range(2*i+1):
            print("*",end="")
        print()
    #lower diamond
    for i in range(n-1,-1,-1):#till -1 because it has to print one extra line
        for j in range(n-1-i):
            print(" ",end="")
        for k in range(2*i+1):
            print("*",end="")
        print()
        
diamond(6)
```

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n-1)

def comb(n, k):
    if k == 0 or k == n:
        return 1
    else:
        return int(fact(n) / (fact(k) * fact(n - k)))

def pascal(n):
    n += 1
    for i in range(n):
        for j in range(n - i):
            print(' ', end='')
        for j in range(i + 1):
            print(comb(i, j), end=' ')
        print('\n')

x = int(input('Eneter the number of rows of the pascal triangle you want to see : '))
pascal(x)
```
