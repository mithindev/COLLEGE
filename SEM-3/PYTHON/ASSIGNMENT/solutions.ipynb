Q - 4] Create a function in Python that accepts one parameter: a string that’s 
a sentence. This function should return True if any word in that 
sentence contains duplicate letters and False if not.

```python
def count_letter (word, ch):
  count = 0

  for i in word:
    if i == ch:
      count += 1

  return count

def isUnique(word):
  for ch in word:
    if (count_letter(word, ch) > 1):
      return False
  
  return True


def containsDuplicate (sentence):
    words = sentence.split()

    for word in words:
      if (not isUnique(word)):
        return True

    return False

print(containsDuplicate("Hello I am Mithin Dev"))
```

Q - 6] Create a function that takes a string S. The task is to print all unique permutations of the given string in lexicographically sorted order. 

![Alt text](image-2.png)

```python
def permutation (ans, word):
  ans_lst = []

  if word == None or len(word) == 1:
    ans_lst = [ans + word]
    return ans_lst
  
  ch = word[0]

  for i in range(0, len(ans) + 1):
    left = ans[:i]
    right = ans[i:]

    ans_lst += permutation(left + ch + right, word[1:])

  return ans_lst


print(permutation("", "abc"))
```
