```python
def count_letter (word, ch):
  count = 0

  for i in word:
    if i == ch:
      count += 1

  return count

def isUnique(word):
  for ch in word:
    if (count_letter(word, ch) > 1):
      return False
  
  return True


def containsDuplicate (sentence):
    words = sentence.split()

    for word in words:
      if (not isUnique(word)):
        return True

    return False

print(containsDuplicate("Hello I am Mithin Dev"))
```

Q - 6] Create a function that takes a string S. The task is to print all unique permutations of the given string in lexicographically sorted order. 

![Alt text](image-2.png)

```python
def permutation (ans, word):
  
  if len(word) == 0:
    return [ans]
  
  ch = word[0]

  ans_lst = []

  for i in range(0, len(ans) + 1):
    left = ans[:i]
    right = ans[i:]

    ans_lst += permutation(left + ch + right, word[1:])

  return ans_lst

ans = permutation("", "abc")
print(ans)
```

```python
def lcs(text1: str, text2: str) -> int:
        if len(text1) == 0 or len(text2) == 0:
            return 0
        elif text1[len(text1) - 1] == text2[len(text2) - 1]:
            return 1 + lcs(text1[:len(text1) - 1], text2[:len(text2) - 1])
        else:
            return max(lcs(text1[:len(text1)], text2[:len(text2) - 1]), lcs(text1[:len(text1) - 1], text2[:len(text2)]))

print(lcs('abc', 'abcd'))
```

```python
values =	{
  "I": 1,
  "V": 5,
  "X": 10,
  "L": 50,
  "C": 100,
  "D": 500,
  "M": 1000,
}

def roman2int(roman):
  ans = 0
  i = 0

  while i < len(roman):
    t1 = values[roman[i]]

    if (i + 1 < len(roman)):
      t2 = values[roman[i + 1]]

      if t1 >= t2:
        ans += t1
      else:
        ans -= t1
      i += 1

    else:
      ans += t1
      i += 1
  return ans

print(roman2int('IV'))
```

```python
def print_pascal(n):
  outer = []

  if n <= 0:
    return outer

  for i in range (n):
    inner = []

    if (i == 0):
      inner.append(1)
      outer.append(inner)
      
    else:
      for j in range (i + 1):  
        if (j == 0 or j == i):
          inner.append(1)
        else:
          prev = outer[i - 1]
          inner.append(prev[j] + prev[j - 1])

      outer.append(inner)    
  return outer

ans = print_pascal(5)
print(ans)
```
