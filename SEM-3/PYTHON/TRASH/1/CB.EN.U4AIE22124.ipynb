#### 4.)For the given list filter all elements that do not contain e.

```python
import re
def filt(input_list):
    pattern = r"(?!.*e).*$"    
    filtered_strings = list(filter(lambda string: re.match(pattern, string), input_list))    
    return filtered_strings
my_strings = ['Mango', 'Apple', 'Tree', 'Hall']
filtered_result = filt(my_strings)
print(filtered_result)
```

#### 10)For the given string, replace 0xA0 with 0x7F and 0xC0 with 0x1F

```python
def rep(s):
    pat1 = '0xA0'
    pat2 = '0xC0'
    
    res = re.sub(pat1, '0x7F', s)
    res = re.sub(pat2, '0x1F', res)
    
    return res

print(rep('start address: 0xA0, func1 address: 0xC0'))
```

#### 14.)For the given input llist, filter all elements that start with 'den' or end with 'ly'

#### 15.) For the given input list, filter all whole elements 12\nthree irrespective of case.

```python
import re

def chk(s):
    pattern = r'^12\nThree$'  
    return list(filter(lambda x: re.search(pattern, x, re.IGNORECASE), s[1:]))

print(chk(['12\nthree\n', '12\nThree', '12\nthree\n4', '12\nthree']))
```

#### 16.)For the given input list, replace hand with X for all elements that start with hand followed by at least one word character

```python
import re
items = ['handed', 'hand', 'handy', 'unhanded', 'handle', 'hand-2']
pattern = r'hand\w'
res = [re.sub(pattern, 'X', item) for item in items]
print(res)
```
