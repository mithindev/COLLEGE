1. Sort the hyphen seperated string and return it as hyphen seperated string

````python
def hyphensep(words):
    words = words.split("-")
    words.sort()
    string = ""
    for x in words:
        string+= f"{x}-"
    return string[:-1]

def hyphensep_join(words):
    words = words.split("-")
    words.sort()
    return "-".join(words)

word = "Hello-This-Is-Mithin-Dev"
print(hyphensep(word))
print(hyphensep_join(word))
````

2. Generate a list in a range (Containing Squares)

````python
def listgen_square(start,end):
    return [x*x for x in range(start,end+1)]

print(listgen_square(2,10))
````

3. Remove duplicates from a sorted list

````python
def remove_duplicates(ls):
    if len(ls)<=1: 
        return ls
    i = 0
    j = 1
    while j <len(ls):
        if ls[i] == ls[j]:
            ls.pop(j)
        else:
            j+=1
            i+=1
    return ls

ls = [1,1,1,2,3,4,5,5,5,6,6,7,7,8]
print(remove_duplicates(ls))
````

4. Change the position of the vowels

````python
def change_vowels(string):
    vow=[]
    ind=[]
    for x in range(len(string)):
        if string[x] in "aeiouAEIOU":
            vow.append(string[x])
            ind.append(x)
    ind = ind[::-1]
    ls = list(string)
    for x in range(len(ind)):
        ls[ind[x]] = vow[x]
    return "".join(ls)

word = "Hello World"
print(change_vowels(word))
````

```
Given a string and a string dictionary, find the longest string in the dictionary  
that can be formed by deleting some characters of the given string. If there are  
more than one possible results, return the longest word with the smallest  
lexicographical order. If there is no possible result, return the empty string.  
Example:  
Input: d = {"ale", "apple", "monkey", "plea"}  
S = "abpcplea"  
Output: "apple"  
Explanation: After deleting "b", "c" ,"a" S became "apple" which is present in d.  
```

````python
def findLongestWord(s, d):
    d = sorted(d, key=lambda x: (-len(x), x)) 
    for word in d:
        if isSubsequence(word,s):
            return word
    return ""

def isSubsequence(word,s):
        i, j = 0, 0
        while i < len(s) and j < len(word):
            if s[i] == word[j]:
                j += 1
            i += 1
        return j == len(word)

d = {"ale", "apple", "monkey", "plea"}
S = "abpcplea"
result = findLongestWord(S, d)
print(result)
````
