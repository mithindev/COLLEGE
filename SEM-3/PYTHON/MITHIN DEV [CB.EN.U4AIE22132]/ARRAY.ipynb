### Move the negative elements of the array

```python
# Move the negative elements of the array
import array as ar
def neg(arr):
    i = 0
    count = 0
    while i<len(arr) and count<len(arr):
        if arr[i]<0:
            arr[i],arr[len(arr)-1-count] = arr[len(arr)-1-count], arr[i]
            count +=1
        if arr[i]>=0:
            i+=1
    return arr

a = ar.array('i',[-1,2,4,-2,4,-2,-4,0])
print(neg(a))
```

### Duplicates

```python
import array as arr
# Method 1
def duplicates(array):
    ret = set()
    array = sorted(array)
    i = 1
    while i<len(array):
        if array[i] == array[i-1]:
            ret.add(array[i])
        i+=1
    return ret


a = arr.array('i',[1,2,2,3,5,4,2,3])
print(duplicates(a))
```

### Rotate

```python
import array as arr

def rotate(array,times):
    for x in range(times):
        val = array[-1]
        for y in range(len(array)):
            array[len(array)-1-y] = array[len(array)-y-2]
        array[0] = val
    return array

a = arr.array('i',[1,2,3,4,5,6,7])
print(rotate(a,3))
```
