```python
def countTarget (str, target):
    count = 0

    for ch in str:
        if (ch == target):
            count += 1

    return count
```

```python
def isPalindrome (str):
    l = len(str)

    for i in range(l//2):
      if (str[i] != str[l - i - 1]):
       return False
    
    return True
```
