```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

f = open('DNA_string.txt', 'r')
sequence = f.read()
print(sequence)

n_a = 0
n_t = 0
n_c = 0
n_g = 0

def Xn(n_a, n_t, n_c, n_g):
    return (n_a + n_g) - (n_c + n_t)

def Yn(n_a, n_t, n_c, n_g):
    return (n_a + n_c) - (n_g + n_t)

def Zn(n_a, n_t, n_c, n_g):
    return (n_a + n_t) - (n_c + n_g)

arr = []

for ch in sequence:
    if ch == 'A':
        n_a += 1
    elif ch == 'T':
        n_t += 1
    elif ch == 'C':
        n_c += 1
    else:
        n_g += 1

    arr.append([Xn(n_a, n_t, n_c, n_g), Yn(n_a, n_t, n_c, n_g), Zn(n_a, n_t, n_c, n_g)])
    print(f"{ch} -> Xn: {Xn(n_a, n_t, n_c, n_g)}, Yn: {Yn(n_a, n_t, n_c, n_g)}, Zn: {Zn(n_a, n_t, n_c, n_g)}")

# Extract Xn, Yn, and Zn values from arr
xn_values = [item[0] for item in arr]
yn_values = [item[1] for item in arr]
zn_values = [item[2] for item in arr]

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xn_values, yn_values, zn_values, c='b', marker='o')

ax.set_xlabel('Xn')
ax.set_ylabel('Yn')
ax.set_zlabel('Zn')

plt.title('3D Scatter Plot of Xn, Yn, and Zn')
plt.show()
```
