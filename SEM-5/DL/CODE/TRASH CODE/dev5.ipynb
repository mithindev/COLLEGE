```python
import numpy as np
import torch
import matplotlib.pyplot as plt

def f(x):
    return x ** 2

def f_grad(x):
    return 2 * x

def gd(eta, f_grad):
    x = 100.0
    results = [x]
    for i in range(1000):
        x -= eta * f_grad(x)
        results.append(float(x))
    print(f'Final epoch, x: {x:.6f}')
    return results

def show_trace(results, f):
    n = max(abs(min(results)), abs(max(results)))
    xvals = np.linspace(-n, n, 100)
    yvals = f(xvals)
    plt.figure(figsize=(10, 6))
    plt.plot(xvals, yvals, label='Objective function')
    plt.plot(results, f(np.array(results)), color='orange', linestyle='-', marker='o', label='Gradient descent steps')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gradient Descent Trace')
    plt.legend()
    plt.show()

results = gd(0.9, f_grad)
show_trace(results, f)
```
